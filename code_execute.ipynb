## Docker Compose方案

```python
import json
from jupyter_client import BlockingKernelClient
import os
import glob


def get_latest_conn_json(path):
    json_files = glob.glob(f'{path}/conn-*.json')
    json_files.sort(key=lambda x: os.path.getmtime(x), reverse=True)
    if json_files:
        return json_files[0]
    else:
        return None

def get_kernel_client(connection_file):
    client = BlockingKernelClient(connection_file=connection_file)
    client.load_connection_file()

    client.ip = "127.0.0.1"
    port_start = 11111
    # 注意端口顺序
    client.shell_port = port_start
    client.iopub_port = port_start +1
    client.stdin_port = port_start +2
    client.hb_port = port_start +3
    client.control_port = port_start +4
    return client
```

```python
# first round
code = """import pandas as pd

# 读取远程文件
url = 'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
df = pd.read_csv(url)

# 查看前几行数据
df.head()"""
```

```python
# second round
code = "df.variety.value_counts()"
```

```python
stdout = []
stderr = []
result = ""
xerror = None

connection_file = get_latest_conn_json(f"./data/conn/")
kc = get_kernel_client(connection_file=connection_file)
kc.wait_for_ready(timeout=10)
kc.start_channels()
result_msg_id = kc.execute(
    code=code,
    silent=False,
    store_history=True,
    allow_stdin=False,
    stop_on_error=True,
)
try:
    # TODO: interrupt kernel if it takes too long
    while True:
        message = kc.get_iopub_msg(timeout=180)

        print(json.dumps(message, indent=2, default=str))

        assert message["parent_header"]["msg_id"] == result_msg_id
        msg_type = message["msg_type"]
        if msg_type == "status":
            if message["content"]["execution_state"] == "idle":
                break
        elif msg_type == "stream":
            stream_name = message["content"]["name"]
            stream_text = message["content"]["text"]

            if stream_name == "stdout":
                stdout.append(stream_text)
            elif stream_name == "stderr":
                stderr.append(stream_text)
            else:
                assert False, f"Unsupported stream name: {stream_name}"

        elif msg_type == "execute_result":
            execute_result = message["content"]["data"]
            result = execute_result
        elif msg_type == "error":
            error_name = message["content"]["ename"]
            error_value = message["content"]["evalue"]
            error_traceback_lines = message["content"]["traceback"]
            if error_traceback_lines is None:
                error_traceback_lines = [f"{error_name}: {error_value}"]
            error_traceback = "\n".join(error_traceback_lines)
            xerror = error_traceback
        else:
            print(f"Unsupported msg_type: {msg_type}")
            pass
finally:
    kc.stop_channels()

print(stdout)
print(stderr)
print(result)
print(xerror)
```
